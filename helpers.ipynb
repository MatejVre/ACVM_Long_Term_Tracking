```python
import matplotlib.pyplot as plt
```

```python
precisions = [0.594, 0.590, 0.638, 0.576]
recalls = [0.412, 0.443, 0.351, 0.323]
Fs = [0.487, 0.506, 0.453, 0.418]
```

```python
l = range(2, len(precisions)+2)
plt.plot(l, precisions, label="Precision")
plt.plot(l, recalls, label="Recall")
plt.plot(l, Fs, label="F-Score")
plt.xticks(l, [2, 3, 4, 5])
plt.legend()
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.savefig("thresholds.pdf", bbox_inches="tight")
```
